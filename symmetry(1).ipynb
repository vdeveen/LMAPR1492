# Import et téléchargement du CIF

```python
from pymatgen import Structure
from pymatgen import symmetry
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from pymatgen.core.operations import SymmOp
from IPython.display import Image
with MPRester("izD7mJmnjhUOKyWGtZ") as m:

    # Structure for material id
    structure = m.get_structure_by_material_id("mp-30847")
    w = CifWriter(structure)
    w.write_file('mp-30847.cif')
    sGP= SpacegroupAnalyzer(structure)
    
    
```

La structure cristalline étudié est le TiSnPt, qui possède 96 éléments de symétrie différents. Par la suite, nous étudierons 3 de ces élements sur 1 atome différent chacun. 

# 1e Élémént de symétrie :
On a pour la 1e opération  de symétrie étudiée le 2e élément de symétrie : une rotoinversion d'ordre 4 dans la direction $[0 0 \overline{1}]$.<br /> Elle envoie tout atome en position initiale (x,y,z) à la position (-x,y,-z). Elle s'obtient à l'aide d'une matrice associée à la rotation et un vecteur associé à la translation que l'on retrouve ci dessous : <br />
Si l'on applique cette opération de symétrie sur Sn1 aux coordonnées (0, 0, 0.5) on retrouvera les coordonnées (0, 0 ,0 , -0.5)

```python
    print("2ème élément de symétrie (4-bar axis:y,-x,-z):")
    Op1=SymmOp.from_xyz_string("y, -x, -z")
    print(Op1.rotation_matrix)
    print("Vecteur de translation de la symétrie:")
    print(Op1.translation_vector)
    print("Effet sur Sn1 #2:")
    Eff1=Op1.operate((0,0,1/2))
    print(Eff1)
    display(Image(filename='Symmetry1.png'))
    
   
```

# 2e  Opération de symétrie étudiée 
On a pour la 2e opération  de symétrie étudiée le 7e élément de symétrie : une rotation d'ordre 2 dans la direction $[0 0 \overline{1}]$.<br /> Elle envoie tout atome en position initiale (x,y,z) à la position (-x,y,-z). Elle s'obtient à l'aide d'une matrice associée à la rotation et un vecteur associé à la translation que l'on retrouve ci dessous : <br />
Si l'on applique cette opération de symétrie sur Pt2 aux coordonnées (0.75, 0.25, 0.25) on retrouvera les coordonnées (-0.75, 0.25, -0.25)

```python
    print("7ème élément de symétrie (C2 axis:-x,y,-z):")
    Op2=SymmOp.from_xyz_string("-x,y,-z") 
    print("Matrice de transformation de la symétrie:")
    print(Op2.rotation_matrix)
    print("Vecteur de translation de la symétrie:")
    print(Op2.translation_vector)
    print("Effet sur Pt2 #6:")
    Eff2=Op2.operate((0.75,0.25,0.25))
    print(Eff2)
    display(Image(filename='Symmetry2.png'))
 
```

# 3e  Opération de symétrie étudiée 
On a pour la 3e opération  de symétrie étudiée le 14e élément de symétrie : un plan mirroir dans la direction $[1 0 1]$.<br /> Elle envoie tout atome en position initiale (x,y,z) à la position (z,y,x). Elle s'obtient à l'aide d'une matrice associée à la rotation et un vecteur associé à la translation que l'on retrouve ci dessous : <br />
Si l'on applique cette opération de symétrie sur TiO aux coordonnées (0, 0.5, 0.5) on retrouvera les coordonnées (0.5, 0.5, 0)

```python
    print("14ème élément de symétrie (plan mirroir: z,y,x):")
    Op3=SymmOp.from_xyz_string("z,y,x")
    print(Op3.rotation_matrix)
    print("Vecteur de translation de la symétrie:")
    print(Op3.translation_vector)
    print("Effet sur TiO #12:")
    Eff3=Op3.operate((0,0.5,0.5))
    print(Eff3)
    display(Image(filename='Symmetry3.png'))
```
